```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename1 = 'data/total_cases.csv'
filename2 = 'data/total_deaths.csv'
```

```python
dfc = pd.read_csv(filename1, parse_dates=['date'], index_col = ['date'])
dfd = pd.read_csv(filename2, parse_dates=['date'], index_col = ['date'])   
```

```python
dfd = dfd.fillna(0) # fill NaN by zero
dfc = dfc.fillna(0)
```

```python
for col in dfc.columns: 
    print(col) 
```

```python
country=['India','Maldives','Sri Lanka','Pakistan','Bhutan','Bangladesh','Nepal']
```

```python
dfsc = pd.DataFrame(columns=['India','Maldives','Sri Lanka','Pakistan','Bhutan','Bangladesh','Nepal'])
dfsd =pd.DataFrame(columns=['India','Maldives','Sri Lanka','Pakistan','Bhutan','Bangladesh','Nepal'])
```

```python
dfsc
```

```python
for i in range(len(country)):
    dfsc[country[i]]=dfc[country[i]]
    dfsd[country[i]]=dfd[country[i]]
```

```python
dfsc
```

```python
dfsc.loc['2020-09-29':]
```

```python
dfsd.loc['2020-09-29':]
```

```python
plt.figure(figsize = [12,10])
plt.suptitle('COVID 19 in SARC', fontsize = 15)
for i in range(len(country)):
    plt.subplot(3,3,i+1)
    plt.plot(dfsc.index,dfsc[country[i]],'g-', linewidth = 3 )
    plt.xlabel('Date', fontsize = 12)
    plt.ylabel('Total cases', fontsize = 12)
    plt.xticks(rotation=45)
    plt.title(country[i])
    plt.subplots_adjust(wspace = 0.5, hspace = 0.8)
```

```python
plt.figure(figsize = [12,10])
plt.suptitle('COVID 19 in SARC', fontsize = 15)
for i in range(len(country)):
    plt.subplot(3,3,i+1)
    plt.plot(dfsd.index,dfsd[country[i]],'b-', linewidth = 3,label ='Total deaths' )
    plt.xlabel('Date', fontsize = 12)
    plt.ylabel('Total deaths', fontsize = 12)
    plt.xticks(rotation=45)
    plt.title(country[i])
    plt.subplots_adjust(wspace = 0.5, hspace = 0.8)
```
